```python
# Get pokemon-color
import requests
import time

# Get pokemon-color
from requests import get

## create obj_list
color_name = []


for i in range(10):
    url = f"https://pokeapi.co/api/v2/pokemon-color/{i+1}"
    resp = requests.get(url)
    #check request
    if resp.status_code == 200:
      json_data = resp.json()
      color_name.append(json_data["name"])
      time.sleep(1)
    else:
      print("error")
```

```python
print(color_name)
```
